# Negative feedback are feedback with 'reviews_rating' of 1. Plot a Word Cloud and please also add the chart to the ppt. What is the most frequently occurring word in the negative feedback word cloud of “reviews_text”?

```python
neg_phrases = Amazon_Reviews[ Amazon_Reviews.reviews_rating == 1]
```

```python
neg_phrases.head()
```

```python
neg_words = []
for t in neg_phrases.reviews_text:  # Go to the "reviews_text" column of "neg_phrases" dataframe
    neg_words.append(t)
neg_words[:4]
```

```python
neg_text = pd.Series(neg_words).str.cat(sep = ' ')
```

```python
neg_text[:50]   #'seems', 'VBZ'
```

```python
print(pos_tag(word_tokenize(neg_text[:50])))
```

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud
```

```python
import sys
print(sys.executable)
```

```python
wordcloud = WordCloud(width = 1600, height = 800, max_font_size = 200).generate(neg_text)
```

```python
plt.figure(figsize=(10,10))
plt.imshow(wordcloud)
plt.axis("off")
```

# We made a dataframe which had positive phrases 'reviews_rating' of 1 and primaryCategories as "Electronics." We converted the negative “reviews_title” as a text string.
